```python
import pandas as pd
df=pd.read_csv("/kaggle/input/housing-prices-dataset/Housing.csv")
df.head(2)
```

```python
# Bar charts are seen as common visual tool for displaying a single categoorical variable 
import matplotlib.pyplot as plt
bedroom_count=df['bedrooms'].value_counts().sort_index()
plt.bar(bedroom_count.index,bedroom_count.values)
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.title('Distribution of bedrooms')
plt.show()
```

```python
bathrooms_count=df['bathrooms'].value_counts().sort_index()
bathrooms_count
plt.bar(bathrooms_count.index,bathrooms_count.values)
plt.xlabel('Number of bathrooms')
plt.ylabel('Count')
plt.title('Distribution of bathrooms')
# Set the x-axis ticks with one-unit difference
plt.xticks(range(min(bathrooms_count.index), max(bathrooms_count.index)+1))
# Display the chart
plt.show()
```

```python
stories_count=df['stories'].value_counts().sort_index()
plt.bar(stories_count.index,stories_count.values)
plt.xlabel('Number of stories')
plt.ylabel('Count')
plt.title('Distribution of Stories')
plt.xticks(range(min(stories_count.index),max(stories_count.index)+1))
plt.show()
```

```python
parking_count=df['parking'].value_counts().sort_index()
plt.bar(parking_count.index,parking_count.values)
plt.xlabel('Number of Parking spots')
plt.ylabel('Count')
plt.title('Distribution of parking spots')
plt.xticks(range(min(parking_count.index),max(parking_count.index)+1))
plt.show()
```

```python
import seaborn as sns
correlation_matrix=df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",fmt=".2f",square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.scatter(df['area'],df['price'])
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Scatter Plot: Area vs Price')
plt.show()
```

### Exploring two or more variables, familiar estimators like mean and variance look at variables one at a time (univariate analysis). Correlation analysis compares two variables (bivariate analysis). Now lets have additional estimates and plots and at more than two variables (multivariate analysis)

contingency table, hexaginal binning, contour plot, violin plot are the key terms for exploring two or mor variables

```python
#scatter plot is not suitable when the data is dense, so we need diffrent ways to visualize the relationship

plt.hexbin(df['area'],df['price'],gridsize=15,cmap='YlGnBu', edgecolors='grey')
plt.xlabel('Area')
plt.ylabel('Price in cr')
plt.title('Hexagonal binning plot: Area vs Price')
plt.colorbar(label='count')
plt.show()
```

# Categorical and Numeric Data

Boxplots are simple way to visually compare the distributions of a numeric variable grouped according to a categorical variable.

```python
import seaborn as sns
sns.boxplot(data=df,x='furnishingstatus', y='price')

plt.xlabel('Furnishing Status')
plt.ylabel('Price')
plt.title('Box Plot: Furnishing Status vs. Price')

plt.show()
```

```python
import seaborn as sns
sns.violinplot(data=df,x='furnishingstatus', y='price')

plt.xlabel('Furnishing Status')
plt.ylabel('Price')
plt.title('Violin Plot: Furnishing Status vs. Price')

plt.show()
```
