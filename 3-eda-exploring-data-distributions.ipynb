```python
import pandas as pd
df=pd.read_csv("/kaggle/input/housing-prices-dataset/Housing.csv")
df.head(2)
```

### The first two EDA's covered shows on how we sum up the data into a single number to describe the location and variability of the data. It is also useful to exlpire how the data is distributed overall.

Box plot, frequency table, Histogram, Density Plot are the key terms used in the exploring of the distribution

```python
df['area'].quantile([0.05,0.25,0.50,0.75,0.90,1.00])
```

```python
# Lets Calculate the components of the box plot 

median=df['area'].median()
q1=df['area'].quantile(0.25)
q3=df['area'].quantile(0.75)
IQR=q3-q1
lower_whisker=q1-1.5*IQR
upper_whisker=q3+1.5*IQR
outliers=df[(df['area']<lower_whisker) | (df['area']>upper_whisker)]['area']

print("Median:", median)
print("Q1:", q1)
print("Q3:", q3)
print("IQR:", IQR)
print("Lower Whisker:", lower_whisker)
print("Upper Whisker:", upper_whisker)
print("Outliers:", outliers)
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['area'])
plt.title('Box plot for the area feature')
plt.xlabel('Column area')
plt.ylabel('Sq ft')

# Show the plot
plt.show()
```
